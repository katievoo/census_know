```python
myAPI = '90341a28f02794c77f6e3abb153a4952a0bb480b'

def subject_pull(code, year):
    url = "https://api.census.gov/data/" +  year + "/acs/acs5/subject?get=NAME," +\
            code + "&for=tract:*&in=state:36&in=county:*&key=" + myAPI
    resp = requests.request('GET', url).content
    subject_data = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
    return subject_data

def detail_pull(code, year):
    url = "https://api.census.gov/data/" +  year + "/acs/acs5/?get=NAME," +\
            code + "&for=tract:*&in=state:36&in=county:*&key=" + myAPI
    resp = requests.request('GET', url).content
    detail_data = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
    return detail_data

oops_list = []
error_list = []

for index, data in codes.iterrows():
    vtype = data[1]
    code = data[0]
    if vtype == 'subject':
        df = subject_pull(code, '2017')
        print(code)
        try: 
            avg = df[code].mean()
        except: 
            print('error')
            error_list.append(code)
        if avg < 0:
                oops_list.append(code)
        elif avg > 0: 
                pass
    elif vtype == 'detail':
        df = detail_pull(code, '2017')
        print(code)
        try: 
            avg = df[code].mean()
        except: 
            print('error')
            error_list.append(code)
        if avg < 0:
                oops_list.append(code)
        elif avg > 0: 
                pass
```
